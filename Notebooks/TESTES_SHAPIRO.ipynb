```python
dados = {
    'nome': ['Jaoa', 'Mayria', 'Kakuete'],
    'nota': [10.0, 1.0, 8.5],
    'mensalidade': [1700.0, 2500.0, 1500.0],
    'idade': [20, 30, 40],
    'sexo': ['M', 'F', 'M'],
}
```

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.DataFrame(dados)
df
```

```python
def normalizar(x):
    return (x - np.min(x)) / (np.max(x) - np.min(x))

def padronizar(x):
    return (x - np.mean(x))/np.std(x)
```

```python
normalizar(df.mensalidade)
```

```python
padronizar(df.mensalidade)
```

```python
def shapiro(dados):
    stat, p_valor = stats.shapiro(dados)
    print("Estaística de teste de Shapiro-Wilk: ", stat)
    print("Valor p: ",p_valor)
    alpha = 0.20
    if p_valor > alpha:
        print("Os dados parecem seguir uma distribuição normal")
        return 1
    else:
        print("Os dados não seguem uma distribuição normal")
        return 0
```

```python
for coluna in ['nota', 'mensalidade', 'idade']:
    print(coluna)
    n = shapiro(df[coluna])
    if n == 1:
        df[coluna]= padronizar(df[coluna])
    else:
        df[coluna]= normalizar(df[coluna])
    print("_____________")
```

```python
# axis = 1 (coluna) / axis = 0 (row)

df.drop('nome', axis=1, inplace=True)
```
